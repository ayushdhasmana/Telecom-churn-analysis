```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
print("importing done")

db =pd.read_csv(r"C:\Users\HP\Downloads\raw data Telecom Customer Churn.csv")
db
```

```python
"""we have 11 new customer in dataset because of that we have blank spaces in dataset so we replace blank spaces to 0 
and then converting totalcharges datatype to float"""
db["TotalCharges"] = db["TotalCharges"].replace(" ",0)
db["TotalCharges"] = db["TotalCharges"].astype("float")
```

```python
# in Senior Citizen column we have 1 and 0 as true and false respectivily so we are changing 1 and 0 to true and false
db["SeniorCitizen"] = db["SeniorCitizen"].replace({"1": "True", "0":"False"})
```

```python
# in Senior Citizen column we have 1 and 0 as yes and no respectivily so we are changing 1 and 0 to yes and no
def convert(value):
    if value == 1:
        return "Yes"
    else:
        return "No"
db["SeniorCitizen"] = db["SeniorCitizen"].apply(convert)
```

```python
ax = sns.countplot(x = "Churn", data = db)
ax.bar_label(ax.containers[0])
plt.title("count of churn customers")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Group and get counts
grouped = db.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Step 2: Convert to % of total per SeniorCitizen group
percent = grouped.div(grouped.sum(axis=1), axis=0) * 100

# Step 3: Plot stacked bar chart manually
ax = percent.plot(kind='bar', stacked=True, figsize=(8,6), colormap='Set2')

# Step 4: Add % labels
for idx, bars in enumerate(ax.containers):
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            ax.text(
                bar.get_x() + bar.get_width() / 2,
                bar.get_y() + height / 2,
                f'{height:.1f}%',
                ha='center', va='center', fontsize=10, color='black'
            )

# Final touches
plt.title("Churn by SeniorCitizen (Stacked %)")
plt.ylabel("Percentage")
plt.xlabel("SeniorCitizen (0 = No, 1 = Yes)")
plt.legend(title="Churn")
plt.tight_layout()
plt.show()
```

as per report greated percentage of people in senior citizen category has churned

```python

# List of columns to plot
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set number of rows and columns for subplots
rows = 3
cols_per_row = 3
# Create subplots
fig, axes = plt.subplots(rows, cols_per_row, figsize=(18, 12))
axes = axes.flatten()  # Make it easy to index

# Plot countplot for each column
for idx, col in enumerate(cols):
    ax = axes[idx]
    sns.countplot(x=col, data=db, ax=ax, palette='Set2', hue = "Churn")
    ax.set_title(f'Count of {col}')
    ax.tick_params(axis='x', rotation=45)

# Remove unused axes if any
for i in range(len(cols), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

The majority customers who do not churn tend to have services like PhoneService, InternetService (particolany DSL), and OnlineSecurity enabled. For services like OnlineBackup, TechSupport, and streamingTV, churn rates are noticeably higher when these services are not used or are unavailable.
